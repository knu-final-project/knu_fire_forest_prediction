```python
from sklearn.linear_model import LogisticRegression
import pandas as pd

fire = pd.read_csv('day_weather_firecount.csv', encoding = 'euc-kr')
fire.head(50)
```

```python
fire_ml = fire[['area_code','date','max_temp','day_rain','max_wind','min_humidity','counts']]
```

```python
features = fire_ml[['max_temp','day_rain','max_wind','min_humidity']]
counts = fire_ml[['counts']]
```

```python
from sklearn.model_selection import train_test_split

train_features, test_features, train_labels, test_labels = train_test_split(features, counts , test_size=0.33, random_state=321)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

train_features = scaler.fit_transform(train_features)
test_features = scaler.fit_transform(test_features)
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression()

model.fit(train_features, train_labels)
```

```python
model.score(train_features, train_labels)
```

```python
model.score(test_features, test_labels)
```
