```python
import googlemaps
import pandas as pd

my_key = "#"
maps = googlemaps.Client(key=my_key)  # my key값 입력
lat = []  #위도
lng = []  #경도
```

```python
csv = pd.read_csv('forest_fires_preprocessing.csv',encoding='utf-8-sig')
print(csv.head())
```

```python
type(csv['ADDRESS'])
```

```python
places = []
length = len(csv['ADDRESS'])

for i in range(length) :
    places.append(csv['ADDRESS'][i])
    
#print(place)
```

```python
i=0
for place in places:   
    i = i + 1
    try:
        print("%d번 인덱스에서 %s의 위치를 찾고있습니다"%(i, place))
        geo_location = maps.geocode(place)[0].get('geometry')
        lat.append(geo_location['location']['lat'])
        lng.append(geo_location['location']['lng'])
        

    except:
        lat.append('')
        lng.append('')
        print("%d번 인덱스 위치를 찾는데 실패했습니다."%(i))
```

```python
for i in range(length) :
    csv['lat'] = pd.DataFrame (lat, columns = ['lat'])
    csv['lng'] = pd.DataFrame (lng, columns = ['lng'])
    
```

```python
print('위도 length :' , len(lat), '경도 length :',len(lng))
```
